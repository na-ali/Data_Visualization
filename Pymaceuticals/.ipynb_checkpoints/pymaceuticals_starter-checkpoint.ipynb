```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 

# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean_by = combine_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median_by= combine_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance_by= combine_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
std_by= combine_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem_by = combine_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

# Assemble the resulting series into a single summary DataFrame.
summary_statistics = pd.DataFrame({'Mean Tumor Volume':mean_by,
                                    'Median Tumor Volume':median_by,
                                    'Tumor Volume Variance':variance_by,
                                    'Tumor Volume Std. Dev':std_by,
                                  'Tumor Volume Std. Err.':sem_by})

summary_statistics
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line


adv_summary_statistics = combine_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

adv_summary_statistics
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

specified_regimens  = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
filtered_df = combine_df_data[combine_df_data['Drug Regimen'].isin(specified_regimens)]

# Start by getting the last (greatest) timepoint for each mouse

max_timepoints = filtered_df.groupby('Mouse ID')['Timepoint'].max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint

final_tumor_volumes = pd.merge(max_timepoints, combine_df_data, on=['Mouse ID', 'Timepoint'], how='left')
```

```python
# Put treatments into a list for a loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create an empty list to fill with tumor volume data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers.

for treatment in treatments:
    treatment_df = final_tumor_volumes.loc[final_tumor_volumes['Drug Regimen'] == treatment, 'Tumor Volume (mm3)']
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_vol_data.append(treatment_df)
    
    # add subset
    quartiles = treatment_df.quantile([0.25, 0.5, 0.75])
    lower_quartile = quartiles[0.25]
    upper_quartile = quartiles[0.75]
    iqr = upper_quartile - lower_quartile
    lower_bound = lower_quartile - 1.5 * iqr
    upper_bound = upper_quartile + 1.5 * iqr
    
    # Determine outliers using upper and lower bounds
    outliers = treatment_df[(treatment_df < lower_bound) | (treatment_df > upper_bound)]
    
    print(f"{treatment}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

fig1, ax1 =plt.subplots() 
ax1.set_ylabel('Tumor Volume (mm3)') 
ax1.boxplot(tumor_vol_data, labels=treatments) 

 
plt.show()  
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

treatment_of_capomulin = combine_df[combine_df['Drug Regimen']=='Capomulin']

average_tumor_volume = treatment_of_capomulin.groupby(['Mouse ID'])[['Weight (g)','Tumor Volume (mm3)']].mean()

plt.figure(figsize=(8, 6))
plt.scatter(average_tumor_volume['Weight (g)'], average_tumor_volume['Tumor Volume (mm3)'], color='blue')

plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.ylim(34,46.5)
plt.show()

```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

correlation_coefficient, _ = st.pearsonr(average_tumor_volume['Weight (g)'], average_tumor_volume['Tumor Volume (mm3)'])
print(f'The correlation between mouse weight and the average tumor volume is {round(correlation_coefficient, 2)}.')

(slope, intercept, r_value, _,_) =st.linregress(average_tumor_volume['Weight (g)'], average_tumor_volume['Tumor Volume (mm3)'])

regression_value = average_tumor_volume['Weight (g)']* slope + intercept

plt.figure(figsize=(8, 6))
plt.scatter(average_tumor_volume['Weight (g)'], average_tumor_volume['Tumor Volume (mm3)'], color='blue')
plt.plot(average_tumor_volume['Weight (g)'], regression_value, 'r-')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.ylim(34,46.5)

plt.show()

```
